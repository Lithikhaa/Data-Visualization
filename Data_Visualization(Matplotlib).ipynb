***Subplot --> (2,2,1)   2rows , 2columns , 1position.  Subplot allows you to insert multiple plots on a grid within a single figure***


```python
plt.figure(figsize = (8,8))

plt.subplot(2,2, 1)
plt.scatter(study_hours, marks)

plt.subplot(2,2, 2)
plt.plot(study_hours, marks, 'r--')

plt.subplot(2,2, 3)
plt.hist(marks)

plt.subplot(2,2, 4)
plt.plot(study_hours, marks, 'r-')
plt.plot(study_hours, marks, 'bo')



plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(4,4))

# First subplot in the 2x2 grid
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Plot 1')

# Second subplot in the 2x2 grid
plt.subplot(2, 2, 2)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title('Plot 2')


# Adjust layout to prevent overlap
plt.tight_layout()

# Show the figure with all subplots
plt.show()
```


***Pie Chart***

***Let's say you have a group of students and you want to see how many scored within certain ranges on a test:***

***Scores: 0-10, 11-20, 21-30, etc***.

***Counts: Number of students who scored within each range***.

***Visualization***:
*If you have 10 students, and their scores are 5, 12, 19, 22, 24, 27, 35, 45, 48, 52*

***You could divide the scores into bins like this***

*0-10: 1 student*

*11-20: 2 students*

*21-30: 3 students*

*31-40: 1 student*

*41-50: 2 students*

*51-60: 1 student*

```python
# Sample data: test scores of students
scores = [5, 12, 19, 22, 24, 27, 35, 45, 48, 52]

plt.hist(scores, bins=6, edgecolor='black')

plt.xlabel('Score Ranges')
plt.ylabel('Number of Students')
plt.title('Distribution of Test Scores')

plt.show()
```

***Bar Chart***

***Bar graphs are used to compare things between different groups or to track changes over time.***

***Example:
Imagine you want to show how many apples, bananas, and oranges you have***

***Categories: Apples, Bananas, Oranges***

***Values: The number of each fruit***

```python
# Data
fruits = ['Apples', 'Bananas', 'Oranges']
values = [5, 3, 7]

plt.bar(fruits, values)

plt.xlabel('Fruits')
plt.ylabel('Number of values')
plt.title('Favorite Fruits')

plt.show()
```

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks = [89, 90, 45, 78, 99]


plt.bar(subjects, marks)
plt.show()
```

```python
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.bar(subjects, marks, color = colors)
plt.show()
```

```python
plt.bar(subjects, marks, color = colors, width = 0.6)
plt.show()
```

```python
plt.bar(subjects, marks, color = colors, edgecolor = 'white', linewidth = 4)
plt.show()
```

```python
plt.bar(subjects, marks, color = colors, edgecolor = 'white', linewidth = 4, linestyle = '--')
plt.show()
```

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks1 = [89, 90, 45, 78, 99]
marks2 = [78, 56, 34, 90, 12]




plt.figure(figsize = (8,8))

plt.bar(subjects, marks1)
plt.bar(subjects, marks2)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()
```
