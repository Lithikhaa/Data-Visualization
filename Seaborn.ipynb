#***Seaborn***
***Seaborn focused on making statistical graphics***.

***Easier to create aesthetically pleasing and informative statistical graphics with less code***.

***Specialized in statistical plots (e.g., heatmaps, violin plots, pair plots, box plots)***


#***Comparsion Between Matplotlib and seaborn***

| Feature               | Matplotlib                            | Seaborn                               |
|-----------------------|---------------------------------------|---------------------------------------|
| **Purpose**           | General-purpose plotting library      | Statistical data visualization        |
| **Flexibility**       | Highly customizable                   | Built-in themes for ease of use       |
| **Types of Plots**    | Wide range of plots (line, scatter, bar, etc.) | Statistical plots (heatmaps, pair plots, etc.) |
| **Ease of Use**       | Requires more code to customize       | Easier to create attractive plots     |
| **Complexity**        | More complex and detailed             | Simpler and more readable             |



```python
import seaborn as sns
sns .__version__
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/student_data.csv')
df
```

```python
df.columns
```

```python
sns.relplot(x= 'Hours',y = 'Marks', hue = 'Age', data = df)
```

```python
sns.relplot(x= 'Hours',y = 'Marks', hue = 'Age', data = df,kind = 'line',style = 'Internet')
```

```python
sns.relplot(x='Hours',y='Marks',hue = 'Age',data = df,col='Internet')
```

```python
sns.relplot(x='Hours',y='Marks',hue = 'Internet',data = df,col='Age')
```

without giving multiple = 'stack' --> it also visualize with null values that is represented as gray color in the below graph

```python
sns.displot(df,x = 'Age',binwidth = 0.5,hue = 'Internet')
```

with multiple = 'stack' --> it ignores the null values and visualize based one stacked columns('Internet'-->0,1)

this graph shows that, (age = 15) only 0 - 10 people have use internet
and other 10 - 27 people not use internet

```python
sns.displot(df,x = 'Age',binwidth = 0.5,hue = 'Internet',multiple = 'stack')
```

```python
sns.displot(df,x = 'Age',bins =[18,19,20])
```

```python
sns.displot(df,x = 'Marks',col = 'Internet')
```

```python
sns.displot(df,x = 'Age',y = 'Marks',hue = 'Internet')
```

```python
sns.displot(df,x = 'Age',y = 'Marks',hue = 'Internet',kind = 'kde')
```

```python
sns.pairplot(df)
```

```python
sns.catplot(x = 'Age',y = 'Marks',hue = 'Internet',data = df)
```

```python
sns.catplot(x = 'Age',y = 'Marks',hue = 'Hours',data = df)
```

```python
sns.catplot(x = 'Age',y = 'Marks',kind = 'box',hue = 'Internet',data = df)
```

*Regression plot is used for predicting the next thing*

*Eg - if he/she learn for 10 hours! what mark will be scored?*

```python
sns.regplot(x = 'Hours',y = 'Marks',data = df)
```

```python
sns.lmplot(x = 'Marks',y = 'Hours',data = df,hue = 'Internet',markers = ['o','x'],palette = 'Set1')
```

```python
sns.lmplot(x = 'Marks',y = 'Hours',data = df,hue = 'Internet',markers = ['o','x'],palette = 'Set2')
```
